## Reading data, showing columns, and showing first 10 lines

```python
# Reading data from csv
data_by_rows = []
with open('data/arbres_grenoble_epsg4326.csv') as file:
    data_reader = csv.reader(file)
    data_by_rows = list(data_reader)
columns = data_by_rows[0]
data_by_rows.pop(0)

#Showing column names and index
for index in range(len(columns)):
    print(index, "\t", columns[index])

# Showing first 10 lines
print("\nLes 10 premières lignes du fichier csv :\n")
for i in range(10):
    print(data_by_rows[i])
```

## Années de plantation

```python
# Année de plantation
annee_plantation = list()
for row in data_by_rows:
    annee_plantation.append(row[19])

# Shows first and last 50
print("First 50 lines")
print(annee_plantation[0:50])
print("Last 50 lines")
print(annee_plantation[len(annee_plantation)-51:len(annee_plantation)])
```

## Pour combien d'arbres manque-t-il la date de plantation ?

```python
# Méthode 1
n_annee_plantation_manquante = annee_plantation.count('')
print(n_annee_plantation_manquante)
# Méthode 2
n_annee_plantation_manquante2 = [len(annee)==4 for annee in annee_plantation].count(False)
print(n_annee_plantation_manquante2)

```

## La plus ancienne année de plantation recensée dans ce dataset ?

```python

annee_plantation_int = [int(annee) for annee in annee_plantation if annee!='']
annee_min = min(annee_plantation_int)
print(annee_min)

annee_max = max(annee_plantation_int)
print(annee_max)
```

## Combien d'arbres ont été plantés année par années ?

```python
count_annee = dict()
for annee in range(annee_min, annee_max+1):
    count_annee[str(annee)] = annee_plantation_int.count(annee)

print(count_annee)
plt.hist(x=annee_plantation_int, bins=annee_max-annee_min+1)
plt.xlabel('Année de plantation')
plt.ylabel('Count')
plt.show()
```

## Combien d'arbres plantés en omyenne chaque année ?

```python
avg_plantation_par_an = np.mean(list(count_annee.values()))
print(avg_plantation_par_an)
# Vérification que moyenne * nombre d'années = nombre d'arbres plantés
print(avg_plantation_par_an*(annee_max-annee_min+1) == sum(list(count_annee.values())))
```

## Stocker dans une liste de tuples l'année de plantatino et le nombre d'arbres plantés

```python
plantation_par_an = list()
for annee in count_annee.keys():
    plantation_par_an.append((int(annee), count_annee[annee]))

print(plantation_par_an)
```

## Quel Maire a planté le plus d'arbres à Grenoble ?

```python
# Récupération données maires de Grenoble
with open('data/maires_cleaned.csv') as file:
    maires_reader = csv.reader(file)
    data_maires = dict(maires_reader)
data_maires.pop("annee")
#print(data_maires)

# Count d'arbres par maires
arbres_par_maire = dict()
for annee, plantations in plantation_par_an:
    current_maire = data_maires[str(annee)]
    if current_maire not in arbres_par_maire.keys():
        arbres_par_maire[current_maire] = plantations
    else:
        arbres_par_maire[current_maire] += plantations

#print(arbres_par_maire)
print("Nombre d'arbres plantés :")
print("-------------------------------------")
arbres_par_maire = dict(sorted(arbres_par_maire.items(), key=lambda nb:nb[1], reverse=True))
for maire in arbres_par_maire.keys():
    print(maire, arbres_par_maire[maire])
```

## Quel maire a planté le plus d'arbres par année de mandat ?

```python
annees_de_mandat = dict()
for annee in data_maires.keys():
    maire = data_maires[annee]
    if maire not in annees_de_mandat.keys():
        annees_de_mandat[maire] = 1
    else:
        annees_de_mandat[maire] += 1

print("Années de mandat :")
print("-------------------------------------")

for maire in annees_de_mandat.keys():
    print(maire, annees_de_mandat[maire])
print()

arbres_par_annee_de_mandat = dict()
for maire in annees_de_mandat.keys():
    arbres_par_annee_de_mandat[maire] = int(round(arbres_par_maire[maire]/annees_de_mandat[maire], 0))

print("Arbres plantés par années de mandat :")
print("-------------------------------------")
arbres_par_annee_de_mandat = dict(sorted(arbres_par_annee_de_mandat.items(), key=lambda nb:nb[1], reverse=True))
for maire in arbres_par_annee_de_mandat.keys():
    print(maire, arbres_par_annee_de_mandat[maire])

```
