```python
# Nombre d'espèces (18)
#print(len(espece))

# Récupération des années de plantation
annee_plantation = list()
for row in data_by_rows:
    annee_plantation.append(row[19])
annee_plantation_int = [int(annee) for annee in annee_plantation if annee!='']
annee_min = min(annee_plantation_int)
#print(annee_min)

annee_max = max(annee_plantation_int)
#print(annee_max)

vecteur_annees = range(annee_min, annee_max+1)

# Récupération de l'évolution du nombre d'arbres par genre au fil des années
evolution_genres_botaniques = dict.fromkeys(genres_botaniques,[])
for genre in evolution_genres_botaniques.keys():
    evolution_genres_botaniques[genre] = [0 for i in vecteur_annees]

evolution_genres_botaniques['Non défini'] = [0 for i in vecteur_annees]
for row in data_by_rows:
    annee = row[19]
    if annee!='':
        genre = row[13]
        if genre in evolution_genres_botaniques.keys():
            evolution_genres_botaniques[genre][int(annee)-annee_min] += 1
        else:
            evolution_genres_botaniques['Non défini'][int(annee)-annee_min] += 1

# Récupération de l'évolution du nombre d'arbres par espèce au fil des années
evolution_especes = dict.fromkeys(especes,[])
for espece in evolution_especes.keys():
    evolution_especes[espece] = [0 for i in vecteur_annees]

evolution_especes['Non défini'] = [0 for i in vecteur_annees]
for row in data_by_rows:
    annee = row[19]
    if annee!='':
        genre = row[13]
        espece = row[14]
        current_espece = genre + ' ' + espece
        if current_espece in evolution_especes.keys():
            evolution_especes[current_espece][int(annee)-annee_min] += 1
        else:
            evolution_especes['Non défini'][int(annee)-annee_min] += 1


# Nombre d'arbres plantés par an
arbres_plantes_par_an = [0 for i in vecteur_annees]
for annee in vecteur_annees:
    current_tot = 0
    for espece in evolution_especes.keys():
        current_tot += evolution_especes[espece][int(annee)-annee_min]
    arbres_plantes_par_an[int(annee)-annee_min] = current_tot
#print(arbres_plantes_par_an)

# Proportion de genres botaniques arbres plantés chaque année
evolution_genres_botaniques_proportions = dict(evolution_genres_botaniques)
for genre in evolution_genres_botaniques_proportions.keys():
    for i in range(len(evolution_genres_botaniques_proportions[genre])):
        evolution_genres_botaniques_proportions[genre][i] = evolution_genres_botaniques_proportions[genre][i]/max(1,arbres_plantes_par_an[i])

# Proportion d'espèces arbres plantés chaque année
evolution_especes_proportion = dict(evolution_especes)
for espece in evolution_especes_proportion.keys():
    for i in range(len(evolution_especes_proportion[espece])):
        evolution_especes_proportion[espece][i] = evolution_especes_proportion[espece][i]/max(1,arbres_plantes_par_an[i])

# Nombre d'especes différentes plantées par année
nombre_especes_par_an = [0 for i in vecteur_annees]
for espece in evolution_especes.keys():
    for annee in vecteur_annees:
        i = annee - annee_min
        if evolution_especes[espece][i]!=0:
            nombre_especes_par_an[i] += 1
plt.plot(vecteur_annees, nombre_especes_par_an)
plt.title("Nombre d'espèces différentes plantées par an")
plt.xlabel("Année")
plt.ylabel("Nombre d'espèces")
plt.show()


#for espece in evolution_especes.keys():
#    plt.plot(vecteur_annees, evolution_especes_proportion[espece], )
#plt.show()

dix_genres_plus_communs = ['Acer','Platanus','Pinus','Tilia','Fraxinus','Prunus','Carpinus','Quercus','Betula','Non défini']
y1 = []
for genre_commun in dix_genres_plus_communs:
    y1.append(evolution_genres_botaniques_proportions[genre_commun])
plt.stackplot(vecteur_annees, y1, labels=dix_genres_plus_communs)
plt.xlim((1964,2024))
plt.ylim((0,1))
plt.title("Proportion des 10 genres botaniques les plus plantées(1964 - 2024)")
plt.xlabel("Année")
plt.ylabel("Proportion")
plt.legend(prop={'size': 6})
plt.show()

# Proportion d'especes par an
dix_especes_plus_communes = ['Platanus acerifolia',
'Non défini',
'Acer pseudoplatanus',
'Pinus nigra',
'Carpinus betulus',
'Acer platanoides',
'Acer campestre',
'Tilia tomentosa',
'Liquidambar styraciflua',
'Fraxinus excelsior']
y2=[]
for espece_commune in dix_especes_plus_communes:
    y2.append(evolution_especes_proportion[espece_commune])
plt.stackplot(vecteur_annees, y2, labels=dix_especes_plus_communes)
plt.xlim((1964,2024))
plt.ylim((0,1))
plt.title("Proportion de chaque espèce plantée par an (1964 - 2024)")
plt.xlabel("Année")
plt.ylabel("Proportion")
plt.legend(prop={'size': 6})
plt.show()
```

#### Indice de shannon

Cet indice donne une idée de la diversité spécifique d'un milieu, c'est-à-dire du nombre d'espèces de ce milieu (richesse spécifique) et de la répartition des individus au sein de ces espèces (équitabilité spécifique).

L'indice est une mesure de l'entropie. Elle est représentée par un nombre réel positif souvent compris entre 0 et 5, mais n'ayant en théorie pas de maximum. Ce nombre est calculé à l'aide d'une fonction d'information inversement proportionnelle à la probabilité d'occurrence d'une observationp.45[1].

« L’entropie peut être entendue comme la surprise moyenne fournie par l’observation d’un échantillon », ainsi que l'explique Eric Marconp.45[1]. Prenant un indice de diversité existant, l'entropie indique la quantité d'information apportée par un échantillon en fonction de la fonction d'information. Par exemple, l'observation d'un individu d'une espèce considérée comme rare a une faible probabilité et apporte plus d'information que l'observation d'un individu d'une espèce communep.45[1]. 

$H'= - \Sigma_{i=1}^{S} \frac{n_i}{N} log_{2}(\frac{n_i}{N})$


- $H'$ : indice de biodiversité de Shannon
- $i$ : une espèce du milieu d’étude
- $S$ : richesse spécifique
- $n_i$ : nombre d'individus pour l'espèce i
- $N$ : effectif total

Il est possible de choisir arbitrairement la base du logarithme et on trouve donc souvent dans la littérature scientifique log ou log de base 2 à la place de ln.

Cet indice permet de quantifier l'hétérogénéité de la biodiversité d'un milieu d'étude[6],[7] et donc d'observer une évolution au cours du temps. Cet indice varie toujours de 0 à ln S (ou log S ou log2 S, selon le choix de la base du logarithme)


Cet indice est l'un des plus connus et des plus utilisés par les spécialistes.

Il existe d'autres méthodes pour effectuer le calcul : l'indice de Simpson et l'indice de Hurlbert.

```python
# En cumulé
def cumule_liste(liste):
    n = len(liste)
    if n==0:
        return []
    elif n==1:
        return liste
    liste_cumulee = [liste[0]]
    if n>1:
        for i in range(1,n):
            liste_cumulee.append(liste_cumulee[i-1]+liste[i])
    return liste_cumulee



evolution_especes = dict.fromkeys(especes,[])
for espece in evolution_especes.keys():
    evolution_especes[espece] = [0 for i in vecteur_annees]

evolution_especes['Non défini'] = [0 for i in vecteur_annees]
for row in data_by_rows:
    annee = row[19]
    if annee!='':
        genre = row[13]
        espece = row[14]
        current_espece = genre + ' ' + espece
        if current_espece in evolution_especes.keys():
            evolution_especes[current_espece][int(annee)-annee_min] += 1
        else:
            evolution_especes['Non défini'][int(annee)-annee_min] += 1



evolution_especes_cumulee = dict(evolution_especes)
for espece in evolution_especes_cumulee.keys():
    evolution_especes_cumulee[espece] = cumule_liste(evolution_especes_cumulee[espece] )
print(evolution_especes_cumulee)
liste_indice_shannon_pop_totale = from_population_to_shannon_index(evolution_especes_cumulee)
plt.plot(vecteur_annees, liste_indice_shannon_pop_totale)
plt.xlabel("Année")
plt.ylabel("Indice de Shannon")
plt.title("Indice de Shannon sur les populations cumulées")
plt.xlim((1964, 2024))
plt.show()
```
